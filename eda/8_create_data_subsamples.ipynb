```python
import pandas as pd
import numpy as np
```

```python
def combine_and_shuffle_datasets(df1, df2, sample_percentage):
    
    # Take random samples from each dataset
    sample1 = df1.sample(frac=sample_percentage, replace = False, random_state=42)
    sample2 = df2.sample(frac=sample_percentage, replace = False, random_state=42)
    
    # Concatenate the samples to create a combined dataset
    combined_dataset = pd.concat([sample1, sample2], ignore_index=True)
    
    # Shuffle the combined dataset
    shuffled_dataset = combined_dataset.sample(frac=1, random_state=42).reset_index(drop=True)
    
    return shuffled_dataset
```

```python
toxic = pd.read_csv('data/toxic.csv')
neutral = pd.read_csv('data/neutral.csv')
```

```python
comments_5_df = combine_and_shuffle_datasets(toxic, neutral, .05)
comments_5_df.to_csv('comments_5_df.csv', index = False)

comments_15_df = combine_and_shuffle_datasets(toxic, neutral, .15)
comments_15_df.to_csv('comments_15_df.csv', index = False)

comments_20_df = combine_and_shuffle_datasets(toxic, neutral, .20)
comments_20_df.to_csv('comments_20_df.csv', index = False)

comments_25_df = combine_and_shuffle_datasets(toxic, neutral, .25)
comments_25_df.to_csv('comments_25_df.csv', index = False)

comments_35_df = combine_and_shuffle_datasets(toxic, neutral, .35)
comments_35_df.to_csv('comments_35_df.csv', index = False)

comments_50_df = combine_and_shuffle_datasets(toxic, neutral, .50)
comments_50_df.to_csv('comments_50_df.csv', index = False)
```
